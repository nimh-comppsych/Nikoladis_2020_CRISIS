```python
assert (~dat2.participant_id.isin(tmp_dat.participant_id)).sum() == 0
```

```python
dat2['dup_id2'] = dat2.groupby('hash').hash.apply(lambda x: np.argsort(x))
dat2['hash_dup_id2'] = dat2.hash.astype(str) + '_' + dat2.dup_id2.astype(str)

tmp_dat['dup_id2'] = tmp_dat.groupby('hash').hash.apply(lambda x: np.argsort(x))
tmp_dat['hash_dup_id2'] = tmp_dat.hash.astype(str) + '_' + tmp_dat.dup_id2.astype(str)
```

```python
missing = pd.read_csv(data_dir / 'missing_eth_data.csv')
```

```python
(~no_eth.participant_id.isin(missing.participant_id)).sum()
```

```python
tmp_dat_ebf = tmp_dat.merge(missing, on='participant_id', how='left', suffixes=('', '_backfill'))
tmp_dat_ebf['Ethnicity'] = tmp_dat_ebf.Ethnicity.fillna(tmp_dat_ebf.Ethnicity_backfill)
tmp_dat_ebf = tmp_dat_ebf.drop('Ethnicity_backfill', axis=1)
```

```python
# Some subjects have some responses that are missing ethnicity
# we'll back fill those from their other responses
ids_no_eth = tmp_dat_ebf.loc[tmp_dat_ebf.Ethnicity.isnull(), 'participant_id']
# Check to make sure none of these subjects have multiple other responses
assert tmp_dat_ebf.loc[tmp_dat_ebf.participant_id.isin(ids_no_eth) & tmp_dat_ebf.Ethnicity.notnull()].groupby("participant_id").Ethnicity.nunique().max() == 1
eth_bf2 = tmp_dat_ebf.loc[tmp_dat_ebf.participant_id.isin(ids_no_eth) & tmp_dat_ebf.Ethnicity.notnull(), ['participant_id', 'Ethnicity']].drop_duplicates()
tmp_dat_ebf2 = tmp_dat_ebf.merge(eth_bf2, on='participant_id', how='left', suffixes=('', '_backfill'))
tmp_dat_ebf2['Ethnicity'] = tmp_dat_ebf2.Ethnicity.fillna(tmp_dat_ebf2.Ethnicity_backfill)
tmp_dat_ebf2 = tmp_dat_ebf2.drop('Ethnicity_backfill', axis=1)
```

```python
assert len(tmp_dat_ebf2.groupby('uid').count().query('Ethnicity > 1')) == 0
```

```python
# Finally got an Ethnicity entry for everything
assert tmp_dat_ebf2.Ethnicity.isnull().sum() == 0
```

```python
# Now we'll deal with multiple Etnicity 
# One subject somehow has lowercase "white", so we can fix that
tmp_dat_ebf2.loc[tmp_dat_ebf2.Ethnicity == 'white', 'Ethnicity'] = 'White'

# Get people with multiple ethnicity responses on the un-excluded items
# Grouping by sample because it is possible for the same participant to fill out parent and adult responses
multi_eth_ids = tmp_dat_ebf2.loc[~tmp_dat_ebf2.exclude, :].groupby(["participant_id", 'sample'])[['Ethnicity']].nunique().query('Ethnicity > 1').reset_index().participant_id

# These three individuals all have marked Other as one of their responses, 
# so we'll set their ethnicity for both responses to that
tmp_dat_ebf2.loc[tmp_dat_ebf2.participant_id.isin(multi_eth_ids), 'Ethnicity'] = 'Other'

# Check that it's all fixed
multi_eth_ids = tmp_dat_ebf2.loc[~tmp_dat_ebf2.exclude, :].groupby(["participant_id", 'sample'])[['Ethnicity']].nunique().query('Ethnicity > 1').reset_index().participant_id
assert len(multi_eth_ids) == 0
```

```python
eth_by_sample = tmp_dat_ebf2.loc[~tmp_dat_ebf2.exclude].groupby(['sample', 'Ethnicity'])[['participant_id']].count().reset_index().rename(columns={'participant_id':'n'})
eth_by_sample = eth_by_sample.merge(tmp_dat_ebf2.loc[~tmp_dat_ebf2.exclude].groupby('sample')[['participant_id']].nunique().rename(columns={'participant_id':'sample_n'}), how='left', on='sample')
eth_by_sample['proportion'] = (eth_by_sample.n / eth_by_sample.sample_n) * 100
```

```python
eth_by_country_name = tmp_dat_ebf2.loc[~tmp_dat_ebf2.exclude].groupby(['country_name', 'Ethnicity'])[['participant_id']].count().reset_index().rename(columns={'participant_id':'n'})
eth_by_country_name = eth_by_country_name.merge(tmp_dat_ebf2.loc[~tmp_dat_ebf2.exclude].groupby('country_name')[['participant_id']].nunique().rename(columns={'participant_id':'country_name_n'}), how='left', on='country_name')
eth_by_country_name['proportion'] = (eth_by_country_name.n / eth_by_country_name.country_name_n) * 100
```
